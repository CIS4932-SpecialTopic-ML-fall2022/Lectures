```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.neighbors import NearestNeighbors
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#load Iris dataset: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_iris.html#sklearn.datasets.load_iris
data = load_iris()
print(data.data.shape)
```

```python
x = data.data[:,:2]
y = data.target

x_min, x_max = x[:,0].min()- 0.5, x[:,0].max() + 0.5
y_min, y_max = x[:,1].min() - 0.5, x[:,1].max() + 0.5

#plot training points
plt.scatter(x[:,0],x[:,1],c=y, cmap=plt.cm.Set1, edgecolors='k')
plt.xlabel('Sepal lenght')
plt.ylabel('Sepal wdth')
plt.xlim(x_min,x_max)
plt.ylim(y_min,y_max)
```

```python
#Dive the data into traning and testing sets

X = data.data
Y = data.target

x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.20,random_state=42)

nbr = NearestNeighbors(n_neighbors=3, algorithm='ball_tree')
nbr.fit(x_train)
```

```python
# finding the distances and indices from each training point

distance, indices = nbr.kneighbors(x_test.reshape(-1,4),return_distance=True)
print(distance.shape)
print(distance)
print(indices)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn_cl = KNeighborsClassifier(n_neighbors=3)
knn_cl.fit(x_train,y_train)
```

```python
# check the classification results
classification_results = knn_cl.predict(x_test)
print(classification_results)
```
