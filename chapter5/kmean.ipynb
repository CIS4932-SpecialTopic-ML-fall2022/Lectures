```python
#import the library
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# generate the data using make_blods, a function in sckit-learn to generate systhetic data

samples, true_labels= make_blobs(n_samples=500, centers=5, cluster_std=0.6, random_state=0)
```

```python
#printing five elements for each of the variable retruned
print(samples[:5])
print(true_labels[:5])
```

```python
# plotting dataset
plt.scatter(samples[:,0],samples[:,1], s=20)
plt.show()
```

```python
# running the Kmean algorithm

kmeans = KMeans(init="random",n_clusters=5, n_init=10, max_iter=300, random_state=0)
kmeans.fit(samples)
y_kmeans = kmeans.predict(samples) 
```

```python
plt.scatter(samples[:,0], samples[:,1], c=y_kmeans, s=20, cmap='spring')
centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='blue', s=100, alpha =0.9)
plt.show()
```

```python
kmeans.inertia_
```

```python
# The clusters assigments are stores into one-dimentional Numpy array

print(kmeans.labels_[:5])
print(true_labels[:5])
```

```python
# Runniing the model using different configuration
kmean_kwarg ={"init":"random","n_init":10,"max_iter":1000,"random_state":42}

#create empty list to save the SSE for each k
number_of_cluster = 6
sse = []
for k in range(1,number_of_cluster):
    kmeans = KMeans(n_clusters= k, **kmean_kwarg)
    kmeans.fit(samples)
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1,number_of_cluster),sse)
plt.xticks(range(1,number_of_cluster))
plt.xlabel("Number of Cluster")
plt.ylabel("SEE")
```
